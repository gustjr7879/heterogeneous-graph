추천 시스템의 정의
사용자와 상품으로 구성된 시스템으로, Item 또는 User가 관심을 가질만한 정보를 추천한다.
사용자가 정보를 수집하고 찾는 시간을 줄여주는 것이 목적이다.
특정 사용자(User)가 좋아할 만한 상품(Item)을 추천하거나, 비슷한 상품을 좋아할만한 User를 추천한다.
Push information으로 사용자가 요구하기 전에 작동한다.
사용자(User)로 부터 선호도(Preferences) 여부를 데이터화한다.
Explicit Feedback(평점 등) / Implicit Feedback(사용자의 행동 패턴 등)


사용자 정보와 아이템 정보를 활용해서 사용자와 아이템 사이의 관계를 분석하고 연관관계를 찾고 해당 연관 관계를 점수화한다.

사용자 프로필이란 ? 
사용자 또는 사용자 그룹을 분석 가능한 요소로 프로파일링하여 개인별 추천 또는 사용자 그룹별 추천을 한다.
사용자 구분할 수 있는 정보는 사용자의 고유 정보(나이, 성별, 지역, 학력 등) 그리고 사용자 로그 분석(쿠키, 인터넷 주소, 방문 기록, 클릭 패턴 등 행동정보)로 되어있다.
정보를 수집하기 위해서 직접적인(Explicit) 방법(설문조사, 평가, 피드백)과 간접적인(Implicit)방법 (웹페이지 머무는 시간, 클릭패턴, 검색로그 등)이 있다.

아이템 프로필이란 ?
플랫폼마다 정의하는 아이템의 종류는 다렴, 플랫폼마다 관련 있는 상품 도는 아이템만 추천한다.
추천 아이템 예시로는 영화 음악 등.....


추천 시스템 데이터 종류
Items : 추천할만한 항목으로 가치와 데이터 복잡도에 따라 구분할 수 있다.
low -> 뉴스 책 영화 등
high -> 금융 상품, 직업, 여행 등

User : 여러 목적과 특징을 가지는 집단으로, 행동 패턴 등으로 유저 모델링을 한다.

Transaction : 유저와 아이템간의 상호작용

사용자와 아이템 정보를 분석하여 추천점수를 계산한다.
사용자 또는 아이템 프로필에서 어떤 정보를 사용할지에 따라 추천알고리즘을 결정한다.
사용자 또는 아이템을 추천하기 위해서 각각의 아이템 또는 사용자에 대한 정량화된 기준이 필요하다.
추천 알고리즘의 목적은 점수화(scoring)하는 것이다.

추천 시스템의 구조

추천 시스템의 전체적인 구조는 후보를 생성하는 단계와 랭킹을 매기는 단계로 구분된다.

후보를 생성하는 단계는 수백만개의 아이템 중 사용자의 활동 기록을 기반으로 후보가 될 만한 수백여개의 아이템으로 추리는 과정이고 이러한 후보는 일반적으로 높은 정밀도로 사용자와 관련이 있고 협업 필터링(사용자 간의 유사성)을 통해서만 광범위한 개인화를 제공한다.
리스트에서 최상의 리스트를 제시하기 위해서 recall이 높은 후보 간의 중요성을 구분해야한다.

랭킹을 매기는 단계는 아이템과 사용자를 설명하는 Feature을 사용해서 원하는 목적 함수에 따라 각 아이템에 점수를 할당하여 가장 높은 점수를 받은 아이템이 점수에 따라 순위가 매겨져 사용자에게 표시된다.


추천시스템의 종류
1. Contents-based Recommender System (컨텐츠 기반 추천시스템)
사용자가 과거에 좋아했던 아이템을 파악하고 그 아이템과 비슷한 아이템을 추천한다.
'부산행'에 5점 평점을 준 사용자 -> '멜로' 영화보다는 '호러'영화를 더 선호한다.
사용자가 과거에 접한 아이템이면서 만족한 아이템
사용자가 좋아했던 아이템 중 일부 또는 전체와 비슷한 아이템 선정
선정된 아이템을 유저에게 추천

2. Collaborative filtering Recommender Systems (협업필터링 추천시스템)
유사한 성향 또는 취향(관심사)를 가지는 다른 사용자가 좋아한 아이템을 현재 사용자에게 추천
'부산행'에 5점을 준 2명의 사용자가 서로 비슷할 것이라고 판단함
사용자 A와 사용자 B는 모두 같은 아이템에 대해 비슷한 또는 같은 평가를 내렸다.
이때 사용자 A는 다른 아이템에도 비슷한 호감을 나타낸다.
따라서 사용자 A, B의 성향이 비슷할 것이므로 A가 선호하는 다른 아이템을 B에게 추천한다.

Steps:
1) 사용자-아이템 평가 매트릭스
2) 사용자-사용자 유사성 매트릭스(코사인 유사도)
3) 유사한 사용자 탐색
4) 후보자 생성(유사한 사용자가 접한 아이템들의 ranking을 살핌)
5) 후보자 점수화(유사한 사용자가 가장 좋아하는 아이템 부터 덜 좋아하는 것 까지 순위)
6) 후보 필터링(이미 사용자가 접한 아이템 제거)

Memory-based : 사용자가 아이템을 좋아하거나 평가했는지, 또는 특정 사용자가 항목을 좋아하거나 평가했는지 여부와 같은 사용자 행동을 관찰한다. 전처리 없이 raw-data에 적용한다. 구현이 쉽고 추천 결과를 설명하기 쉽다는 장점이 있다.

1- User-based : 사용자와 유사한 사용자가 구매/좋아요 했다는 사실을 기반으로 사용자에게 추천함
2- Item-based : 이 항목을 좋아한 사용자가 ####도 좋아했습니다. 느낌 사용자는 마음을 바꿀 수 있기 때문에 User-based보다 안정적이고 새로운 사용자가 사이트를 방문한 경우 User-based보다 더 나은 방식이다.

Model-based : 데이터가 아닌 모델을 기반으로 작업 속도를 높이는 추천시스템을 제공하고 더 나은 확장성을 제공한다. 차원 축소가 자주 사용되고 이 접근법의 가장 유명한 유형은 행렬분해이다.
행렬분해는 채워지지 않은 평가 매트릭스(별점과 같은)를 분석하려고 행렬을 곱해서 loss를 줄여나가는 방법이다.



추천시스템의 정확도를 알 수 있는 대표적인 평가지표를 알아보자 !

평점 예측 등에 사용되는 평가지표로는 RMSE(Root Mean Squre Error)가 있다.
평점 등 예측 문제의 추천 성능을 평가할 때 사용하는 지표이고 RMSE가 낮을 수록 추천 알고리즘의 성능이 좋다고 정량적으로 평가 가능하다.

랭킹 추천에 많이 사용되는 평가지표로는 NDCG (Normalized Discounted Cummluative Gain)이 있다.
정보 검색에서 많이 사용했던 지표로 검색엔진, 영상, 음악 등 컨텐츠 랭킹 추천에서 주요 평가지표로 활용한다. Top-k개의 랭킹 리스트를 만들고 더 관심있거나 관련성 높은 아이템 포함 여부를 평가한다.

MAP(Mean Average Precision), Top K Precision/Recall 등 평가 방법 보완
추천 또는 정보검색에서 특정 아이템에 biased한 경우 사용한다.

rel(Relvance Score) : 사용자에게 추천된 각 아이템이 얼마나 관련있는지, 선호하는지 나타내는 점수이다. binary(관련 여부) or complex value(문제에 따라 세분화 된 값)으로 표현한다.
아이템이 추천리스트에 있지만, 아무 액션이 안일어나면 0, 사지는 않았지만 클릭했다면 1, 예전에 구매했으면 2 ... 처럼 사용한다.

CGp(Cumulative Gain) : 상위 아이템 p개에 대해서 동일한 비중으로 합한다.
추천아이템의 순위를 고려하지 않음

DCGp(Discounted Cumulative gain) : 상위 아이템 p개에 대해서 순위에 따라서 다른 비중으로 사용한다.
추천의 길이가 길어질수록 DCG가 증가하는 경향이 있다.

NDCGp : 이 두개를 계산하여 얻을 수 있다. 1에 가까울 수록 우수한 추천시스템이다.
즉 각각의 값을 보완하고 있다. 하지만 관련없는 추천에 대한 패널티가 없다. 또한 누락된 아이템에 대한 패널티가 없다.

```python
#아이템 기반 추천시스템
import pandas as pd

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

print(movies)
print(ratings)
```

```python
print(ratings.head(1))
```

```python
ratings_mat = ratings.pivot_table('rating','userId','movieId')
print(ratings_mat)
#ratings를 관심 분야인 rating, userId, movieId 만 가져옴
```

```python
#평점을 하지 않았다면 NaN값으로 나오고 있음 이를 제거해줘야함
#제거해주기 전에 어떤 영화인지 직관적으로 알게하기 위해서 movieId를 title로 변경해준다
ratings_movie = pd.merge(ratings, movies, on='movieId')
ratings_matrix = ratings_movie.pivot_table("rating", "userId", "title")

ratings_matrix.fillna(0,inplace=True)

#아이템 - 사용자 행렬로 전치해줌
ratings_matrix_T = ratings_matrix.T
print(ratings_matrix_T.head(3))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#아이템 유사도 행렬을 만들어준다. (각 아이템(영화)이 평점을 비슷하게 받으면)
item_sim = cosine_similarity(ratings_matrix_T,ratings_matrix_T)

item_sim_df = pd.DataFrame(item_sim,index=ratings_matrix_T.index,columns=ratings_matrix_T.index)

print(item_sim_df.head(4))
```

```python
topk1 = item_sim_df["Godfather, The (1972)"].sort_values(ascending=False)[1:6]

print(topk1) #Godfather와 유사한 영화 추천해줌
```
